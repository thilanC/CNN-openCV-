```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img=cv2.imread(r"D:\University\Extra Courses\DL workshop\Day 04\Samples\Sample.jpg")
```

```python
plt.imshow(img)
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # convert my color image to gray
```

```python
plt.imshow(gray_image,cmap='gray')
```

```python
kernal=np.array([[-1,-2,-1],
                 [0,0,0],
                 [1,2,1]],np.float)
```

```python
conv=cv2.filter2D(gray_image,-1,kernal)
```

```python
plt.imshow(conv,cmap='gray')
```

```python
kernal2=np.array([[-1,0,1],
                 [-2,0,2],
                 [-1,0,1]],np.float)

conv2=cv2.filter2D(gray_image,-1,kernal2)

plt.imshow(conv2,cmap='gray')
```

```python
kernal3=np.array([[-2,5,3],
                 [0,0,-3],
                 [2,-1,0]],np.float)

conv3=cv2.filter2D(gray_image,-1,kernal3)

plt.imshow(conv3,cmap='gray')
```
